```python
# diffie hellman only shared key generation
# Alice and bob 
# q = prime number 
# find alpha = primitive root (q)
# Alice  = XA BOB = XB
# YA = pow(alpha,XA)%q be sent to Bob
# YB = pow(alpha,XB)%q be sent to Alice

# k1 = pow(YB,XA)%q
# k2 = pow(YA,XB)%q

# k1 has to be equal k2


def primitiveroot(q):
  num1 = []
  for i in range(2,q-1):
    if(len(num1) == q-1):
      return i
    else:
      num1=[]
    for j in range(0,q-1):
      temp = pow(i,j)%q
      if temp in num1:
        break
      else:
        num1.append(temp)
  
def keygeneration(q,XA,XB):
  alpha = primitiveroot(q) - 1
  YA = pow(alpha,XA)%q
  YB = pow(alpha,XB)%q
  k1 = pow(YB,XA)%q
  k2 = pow(YA,XB)%q

  if k1==k2:
    print("ALICE AND BOB SHARE", k1," ------ ",k2)
  else:
    print("error")

keygeneration(23,3,6)
```
