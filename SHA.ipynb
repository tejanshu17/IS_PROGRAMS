```python
import hashlib
string = "tejanshu"

result1= hashlib.sha1(string.encode())
print("sha 1 result",result1.hexdigest())

result2= hashlib.sha256(string.encode())
print("sha 256 result",result2.hexdigest())

result3= hashlib.sha512(string.encode())
print("sha 512 result",result3.hexdigest())
```
